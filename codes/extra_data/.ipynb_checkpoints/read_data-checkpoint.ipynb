```python
for i in file_names:
    print(i)
    fin = open(i + "_full.txt")
    complete_data = fin.read().splitlines()
    fin.close()
    
    # Find position from which the dataset tabular form begins
    pos = complete_data.index("!dataset_table_begin")
    data = [i.split("\t") for i in complete_data[pos+1:-1]]
    df = pd.DataFrame(data[1:], columns=data[0])
    # Remove the last 19 columns
    new_columns = df.columns[:-19]
    PCOS = []
    # Additionally ignore the first 2 and last 1 column
    # Get the mapping - PCOS, Normal, Other == [1, 0, -1]
    for j in new_columns[2:-1]:
        if j in PCOS_mapping[i]: 
            PCOS.append(1)
        elif j in control_mapping[i]: 
            PCOS.append(0)
        else:
            PCOS.append(-1)
    
    # Transpose and add PCOS mapping column
    df = df[new_columns[2:-1]].T 
    df["PCOS"] = PCOS
    df.index.rename('sample_id', inplace=True)
    
    display(df.head())
    print("="*100)
    
    df.to_csv(i + ".csv")
```

```python
# Get the distribution of samples in each dataset
for i in file_names:
    print(i, "\tPCOS:", len(PCOS_mapping[i]), "\tControl:", len(control_mapping[i]))
```

```python
# Set PCA components to 2
N_COMPONENTS = 2
X_complete = np.empty((1,2))
y_complete = np.empty((1,))

classes = ["PCOS", "Normal/Control", "Other"]
colors = {i:j for i,j in zip(classes, ['r','g','b'])}

for i in file_names:
    df = pd.read_csv(i+".csv", index_col=0)
    X = df.drop(["PCOS"], axis=1).to_numpy()
    y = df["PCOS"]
    
    print(i)
    pca = PCA(n_components=N_COMPONENTS)
    
    # Interpolating with zeros
    X[np.isnan(X)] = 0
    pca.fit(X)
    X_pca = pca.transform(X)
    
    X_complete = np.r_[X_complete, X_pca]
    y_complete = np.r_[y_complete, y]
    
    df_pca = pd.DataFrame(X_pca)
    df_pca = df_pca.set_index(df.index)
    df_pca["PCOS"] = y
    df_pca["PCOS"] = df_pca["PCOS"].astype("str")
    df_pca = df_pca.rename(columns={0:"PC1", 1:"PC2"})

    # fig = px.scatter(df_pca, x="PC1", y="PC2", color="PCOS", title=i+" Data, in 2D latent space")
    # fig.show()
    
    fig, ax = plt.subplots(figsize=[12,6])
    for j,k in enumerate([1, 0, -1]):
        ix = np.where(y==k)
        ax.scatter(df_pca["PC1"].iloc[ix], df_pca["PC2"].iloc[ix], c=colors[classes[j]], label=classes[j], alpha=0.8)
    
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid(True)
    plt.title(i+" Data, in 2D latent space")
    ax.legend()
    plt.show()
    
    print("\n", "="*100)
```
