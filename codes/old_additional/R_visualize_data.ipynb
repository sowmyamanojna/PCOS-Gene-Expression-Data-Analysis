## Reconstruction Images

```r
suppressPackageStartupMessages(library(dplyr))
suppressPackageStartupMessages(library(ggplot2))
suppressPackageStartupMessages(library(cowplot))
```

```r
source(file.path("scripts", "util.R"))
```

```r
source(file.path("scripts", "util.R"))

dataset <- "GEO"
base_dir <- file.path("figures", dataset)

geo_recon_cost_df = read.table(file = 'data/reconstruction_PCA.csv', sep = ',', header = TRUE, row.names=NULL)

geo_recon_gg <- plot_reconstruction_loss(geo_recon_cost_df)

geo_path <- file.path(base_dir, paste0("reconstruction_cost_GEO_PCA"))

save_png_pdf(p = geo_recon_gg,
             path_prefix = geo_path,
             height = 120,
             width = 120)

geo_recon_gg
```
