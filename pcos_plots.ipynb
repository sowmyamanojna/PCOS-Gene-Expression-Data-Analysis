```
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("control_pcos_celltype_mapping.csv")
display(df)
```

```
cell_types = list(set(df['cell_type']))
print(cell_types)
cell_type_freqs = [len(df[df['cell_type'] == x]) for x in cell_types]
cell_type_freqs = [(x/len(df))*100 for x in cell_type_freqs]
print(cell_type_freqs)
```

```
fig1, ax1 = plt.subplots()
ax1.pie(cell_type_freqs, labels=cell_types, autopct='%1.1f%%', shadow=True)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Cell types in PCOS dataset')
# plt.show()
plt.savefig('cell_types.png')
```

```
pcos_control_drug = list(set(df['PCOS/Control']))
print(pcos_control_drug)
pcos_control_drug_freqs = [len(df[df['PCOS/Control'] == x]) for x in pcos_control_drug]
pcos_control_drug_freqs = [(x/len(df))*100 for x in pcos_control_drug_freqs]
print(pcos_control_drug_freqs)
```

```
fig2, ax2 = plt.subplots()
ax2.pie(pcos_control_drug_freqs, labels=pcos_control_drug, autopct='%1.1f%%', shadow=True)

ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Types of samples in PCOS dataset')
# plt.show()
plt.savefig('pcos_control_drug.png')
```
